### Train/Test Split

```python
from sklearn.model_selection import train_test_split

X = df.drop("koi_pdisposition", axis=1)
y = df["koi_pdisposition"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

### Transformation and Encoding Pipeline

```python
mapping = {"FALSE POSITIVE": 0, "CANDIDATE": 1}

y_train = y_train.map(mapping)
y_test = y_test.map(mapping)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
import numpy as np

categorical_columns = ['koi_tce_delivname']
numerical_cols = [c for c in X_train.columns if c not in categorical_columns]


categorical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('cat', categorical_pipeline, categorical_columns),
        ('num', 'passthrough', numerical_cols)
    ]
)

X_train_transformed = preprocessor.fit_transform(X_train)
X_test_transformed = preprocessor.transform(X_test)

cat_columns = preprocessor.named_transformers_['cat']['onehot'].get_feature_names_out(categorical_columns)
all_columns = np.concatenate([cat_columns, numerical_cols])

X_train = pd.DataFrame(X_train_transformed, columns=all_columns, index=X_train.index)
X_test = pd.DataFrame(X_test_transformed, columns=all_columns, index=X_test.index)
```

## Model Training and Evaluation

```python
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

def evaluate_model(y_true, y_pred, model_name):
    """
    Calculates and displays evaluation metrics and the confusion matrix for a model.

    Args:
        y_true: True labels.
        y_pred: Labels predicted by the model.
        model_name: Name of the model (string) to be used in the plot title.
    """
    accuracy = accuracy_score(y_true, y_pred)
    report = classification_report(y_true, y_pred)
    confusion = confusion_matrix(y_true, y_pred)

    print(f"--- Model Evaluation: {model_name} ---")
    print(f"\nAccuracy: {accuracy:.4f}\n")
    print("Classification Report:\n", report)
    print("Confusion Matrix:")

    # Plot Confusion Matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(confusion, annot=True, fmt='d', cmap='Blues', 
                xticklabels=['Not Candidate', 'Candidate'], 
                yticklabels=['Not Candidate', 'Candidate'])
    plt.title(f'Confusion Matrix - {model_name}')
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.show()
```

### XGBoost

```python
xgboost_model = XGBClassifier(random_state=42)

xgboost_model.fit(X_train, y_train)
```

```python
xgboost_predictions = xgboost_model.predict(X_test)
evaluate_model(y_test, xgboost_predictions, "XGBoost")
```

### Random Forest Training

```python
random_forest_model = RandomForestClassifier(random_state=42)

random_forest_model.fit(X_train, y_train)
```

```python
random_forest_predictions = random_forest_model.predict(X_test)
evaluate_model(y_test, random_forest_predictions, "Random Forest")
```

### LightGBM Training

```python
lightgbm_model = LGBMClassifier(objective='binary', random_state=42, verbose=-1)

lightgbm_model.fit(X_train, y_train)
```

```python
lightgbm_predictions = lightgbm_model.predict(X_test)
evaluate_model(y_test, lightgbm_predictions, "LightGBM")
```

### Decision Tree

```python
decision_tree_model = DecisionTreeClassifier(random_state=42)

decision_tree_model.fit(X_train, y_train)
```

```python
decision_tree_predictions = decision_tree_model.predict(X_test)
evaluate_model(y_test, decision_tree_predictions, "Decision Tree")
```
